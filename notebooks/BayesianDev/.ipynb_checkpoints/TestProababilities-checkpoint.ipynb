```python

def chunks(ids1, ids2):
    """Yield successive n-sized chunks from data"""
    for id in ids2:
        w, = np.where( ids1==id )
        yield w
def getTruthTable(gal,cat):
    gal2 = gal[gal['True']==True].copy()

    gal2['Pmem'] = 1.

    indices = list(chunks(gal2['CID'],cat['CID']))
    ngals = np.array([np.sum(gal2['Pmem'][idx]) for idx in indices])
    ngalst = np.array([len(gal2['Pmem'][idx]) for idx in indices])

    cat2 = cat.copy()
    cat2['Ngals'] = ngals
    cat2['R200'] = cat['R200_true'][:]
    cat2['Nbkg'] = cat['Nbkg_true'][:]

    return gal2,cat2

def chunks(ids1, ids2):
    """Yield successive n-sized chunks from data"""
    for id in ids2:
        w, = np.where( ids1==id )
        yield w

def splitBins(var):
    nmin = np.nanmin(var)
    n25 = np.percentile(var,25)
    n50 = np.nanmedian(var)
    n75 = np.percentile(var,75)
    nmax = np.max(var)
    
    return np.array([nmin,n25,n50,n75,nmax])

def getIndices(gindices,gkeys,ckeys):
    indicies = np.empty((0),dtype=int)
    indicies_into_cluster = np.empty((0),dtype=int)

    for i in range(ckeys.size):
        idx, = np.where(gkeys==ckeys[i])
        if idx.size>0:
            w2 = np.arange(gindices[idx],gindices[idx+1], 1, dtype=int)
            w = np.full(w2.size,i,dtype=int)

            indicies = np.append(indicies,w2)
            indicies_into_cluster = np.append(indicies_into_cluster,w)

    return indicies,indicies_into_cluster

def zoffset(z,zcls):
    return (z-zcls)/(1+zcls)

def get_delta_color(gal,cat,indices,lcolor='g-r'):
    gal['delta_rs'] = 0.
    color = gal['%s'%lcolor]
    for i,idx in enumerate(indices):
        rs_param = cat['rs_param_%s'%lcolor][i]
        ri = (color[idx]-rs_param[0])#/rs_param[1]
        if np.nanmedian(ri)>0.05: ri += np.nanmedian(ri)
        gal['delta_rs'][idx] = ri

    return gal['delta_rs']

def set_new_variables(gal, gal2, cat, cat2):
    ## get the same clustes
    cat = cat.group_by('CID')
    cidx, cidx2 = getIndices(cat.groups.indices,cat.groups.keys['CID'],cat['CID'])
    cat2 = cat2[cidx2]
    
    gal = gal[np.unique(gal['GID','CID'],return_index=True)[1]] ## only one galaxy in the cluster
    gal = gal.group_by('CID')
    gindices, gkeys = gal.groups.indices, gal.groups.keys['CID']
    gidx, cidx = getIndices(gindices, gkeys, cat['CID'])
    gal = gal[gidx]

    gal['Ngals'] = cat['Ngals_true'][cidx]
    gal['M200'] = cat['M200_true'][cidx]
    gal['Rnorm'] = gal['R']/cat['R200_true'][cidx]
    gal['dmag'] = gal['mag'][:,2]-cat['magLim'][cidx,1]+3


    gal2 = gal2.group_by('CID')
    gidx, cidx = getIndices(gal2.groups.indices,gal2.groups.keys['CID'],cat['CID'])
    gal2 = gal2[gidx]

    gal2['Ngals'] = cat2['Ngals'][cidx]
    gal2['M200'] = cat2['M200_true'][cidx]
    gal2['Rnorm'] = gal2['R']/cat2['R200_true'][cidx]

    gal['z_offset'] = zoffset(gal['z'],gal['redshift'])
    gal2['z_offset'] = zoffset(gal2['z'],gal2['redshift'])
    ga2l = gal2[np.unique(gal2['GID'],return_index=True)[1]]
    
    # lcolor: 0,1,2,3,4
    color_list = ['g-r','g-i','r-i','r-z','i-z']
    color_index = [[0,1],[0,2],[1,2],[1,3],[2,3]]

    for i,pair_idx in enumerate(color_index):
        i0, i1 = pair_idx
        gal[color_list[i]] = gal['mag'][:,i0]-gal['mag'][:,i1]
        gal2[color_list[i]] = gal2['mag'][:,i0]-gal2['mag'][:,i1]

    indices = list(chunks(gal['CID'],cat['CID']))
    indices2 = list(chunks(gal2['CID'],cat['CID']))

    # gal['delta_rs'] = get_delta_color(gal,cat,indices)
    # gal2['delta_rs'] = get_delta_color(gal2,cat,indices2)

    return gal, gal2, cat, cat2
```

```python
sigmoid = lambda t: 1/(1+np.exp(-t))

def doProb(ngals,nbkg,norm,normed=False):
    
    if normed:
        ngals /= ngals.sum()
        nbkg /= nbkg.sum()

    beta = np.sum(nbkg)/np.sum(ngals)
    
    prob = norm*ngals/(norm*ngals+nbkg)
    prob[np.isnan(prob)] = 0.
    
    prob = np.where(prob>1,1.,prob)
    prob = np.where(prob<0.,0.,prob)
    
    return prob

def scalePDFs(pdfs):
    pm = np.max(pdfs, axis=1)

    return np.product(pdfs/pm,axis=1)

def computeProb(keys, pdfs, pdfs_bkg, norm, nbkg, probz):
    prob = np.empty((1,4),dtype=float)

    pdfr, pdfz, pdfc = pdfs
    pdfr_bkg, pdfz_bkg, pdfc_bkg = pdfs_bkg
    
    for i,key_i in enumerate(keys):
        ni,nb = norm[i],nbkg[i]
        ng = ni*nb/np.mean(pdfr_bkg[key_i])
        pdfri, pdfzi, pdfci = pdfr[key_i],pdfz[key_i],pdfc[key_i]
        pdfr_bkg_i, pdfz_bkg_i, pdfc_bkg_i = pdfr_bkg[key_i],pdfz_bkg[key_i],pdfc_bkg[key_i]
        
        pz = probz[key_i]
        pzb= 1-pz
        # pms = np.max(np.array([pdfri,pdfzi,pdfci]))/10
        # pms = np.max(pdfri)/10, np.max(pdfzi)/10, np.max(pdfci)/10

        # pdfri, pdfzi, pdfci = pdfri/pms[0], pdfzi/pms[1], pdfci/pms[2]
        # pdfr_bkg_i, pdfz_bkg_i, pdfc_bkg_i = pdfr_bkg_i/pms[0], pdfz_bkg_i/pms[1], pdfc_bkg_i/pms[2]
        
        # pdfri, pdfzi, pdfci = pdfri/np.sum(pdfri), pdfzi/np.sum(pdfzi), pdfci/np.sum(pdfci)
        # pdfr_bkg_i, pdfz_bkg_i, pdfc_bkg_i = pdfr_bkg_i/np.sum(pdfr_bkg_i), pdfz_bkg_i/np.sum(pdfz_bkg_i), pdfc_bkg_i/np.sum(pdfc_bkg_i)

        # pdfsi = np.array([pdfri, pdfzi, pdfci])
        # pdfs_bkg_i = np.array([pdfr_bkg_i, pdfz_bkg_i, pdfc_bkg_i])

        # normi = np.average(pdfsi,weights=pz,axis=0)/np.product(np.sum(pdfsi,axis=1))
        # norm_bkgi = np.average(pdfs_bkg_i,weights=pzb,axis=0)/np.product(np.sum(pdfs_bkg_i,axis=1))

        Ngals =  pdfri*pdfzi*pdfci*(np.sum(pdfri)+np.sum(pdfzi)+np.sum(pdfci))/(np.sum(pdfri)*np.sum(pdfzi)*np.sum(pdfci))
        Ngals_bkg = pdfr_bkg_i*pdfz_bkg_i*pdfc_bkg_i*( np.sum(pdfr_bkg_i) + np.sum(pdfz_bkg_i) + np.sum(pdfc_bkg_i) )/(np.sum(pdfr_bkg_i)*np.sum(pdfz_bkg_i)*np.sum(pdfc_bkg_i))

        # Ngals =  pdfri*pdfzi*pdfci*(np.sum(pdfri)+np.sum(pdfzi)+np.sum(pdfci))/(np.sum(pdfri)*np.sum(pdfzi)*np.sum(pdfci))
        # Ngals_bkg = pdfr_bkg_i*pdfz_bkg_i*pdfc_bkg_i*( np.sum(pdfr_bkg_i) + np.sum(pdfz_bkg_i) + np.sum(pdfc_bkg_i) )/(np.sum(pdfr_bkg_i)*np.sum(pdfz_bkg_i)*np.sum(pdfc_bkg_i))

        pr = doProb(pdfri,pdfr_bkg_i,ni)
        pz = doProb(pdfzi,pdfz_bkg_i,ni)
        pc = doProb(pdfci,pdfc_bkg_i,ni)

        pmem = doProb(Ngals,Ngals_bkg,ni,normed=False)

        pi = np.array([pr,pz,pc,pmem]).transpose()
        prob = np.vstack([prob,pi])

    return prob[1:,0], prob[1:,1], prob[1:,2], prob[1:,3]

def commonValues(values):
    idx_sort = np.argsort(values)
    sorted_values = values[idx_sort]
    vals, idx_start, count = np.unique(sorted_values, return_counts=True,
                                return_index=True)

    # sets of indices
    res = np.split(idx_sort, idx_start[1:])
    #filter them with respect to their size, keeping only items occurring more than once

    vals = vals[count > 1]
    commonValuesIndicies = [ri for ri in res if ri.size>1]

    return commonValuesIndicies

def computePtaken(g0,pmem):
    ptaken = np.ones_like(pmem,dtype=float)

    ## find common galaxies
    gid = np.array(g0['GID'])#.astype(np.int)
    commonGroups = commonValues(gid)

    for indices in commonGroups:
        pm_group = np.array(pmem[indices])
        pt_group = np.array(ptaken[indices])

        idx_sort = np.argsort(-1*pm_group) ## -1* to reverse order

        pm_group_s = pm_group[idx_sort]
        pt_group_s = pt_group[idx_sort]

        new_pm = 0
        toto = 1
        pm = []
        for i in range(indices.size):
            toto *= (1-new_pm)
            new_pm = toto*pm_group_s[i]
            pt_group_s[i] = toto
            pm.append(new_pm)

        pmem[indices[idx_sort]] = np.array(pm)
        ptaken[indices[idx_sort]] = pt_group_s

    return pmem, ptaken

def lin_reg(X,Y):
    barX=np.mean(X); barY=np.mean(Y)
    XminusbarX=X-barX; YminusbarY=Y-barY
    b1=sum(XminusbarX*YminusbarY)/sum(XminusbarX**2)
    b0=barY-b1*barX
    Yhat=b0+b1*X
    e_i=Y-Yhat
    sse=np.sum(e_i**2)
    ssr=np.sum((Yhat-barY )**2)
    n=len(X)
    MSE=sse/np.float(n-2)

    s_of_yh_hat=np.sqrt(MSE*(1.0/n+(X-barX)**2/sum(XminusbarX**2)))
    W=np.sqrt(2.0*st.f.ppf(0.95,2,n-2))

    cb_upper=Yhat+W*s_of_yh_hat
    cb_lower=Yhat-W*s_of_yh_hat
    idx=np.argsort(X)

    return {'Yhat':Yhat,'b0':b0,'b1':b1,'cb_u':cb_upper[idx], 'cb_l': cb_lower[idx]}
```

```python
def interpData(x,y,x_new):
#     out = np.empty(x_new.shape, dtype=y.dtype)
    out = interp1d(x, y, kind='linear', fill_value='extrapolate')
    # yint = interp1d(x,y,kind='linear',fill_value='extrapolate')
    return out(x_new)


def get_hdf_pdfs(pdf,ggroup,mode='color',color_idx=[0,1,2,3,4]):
    if mode=='color':
        pdf_new = np.empty_like(ggroup['z'])

        for ix in color_idx:
            color = ggroup[getColorLabel(ix)]
            pdf_i = interpData(pdf[:,0],pdf[:,ix+1],color)
            pdf_new = np.c_[pdf_new,pdf_i]
        return pdf_new[:,1:]

def getColorLabel(color_number):
    color_list = ['g-r','g-i','r-i','r-z','i-z']
    return color_list[color_number]

def open_hdf_get_pdfs(gal,hdf,keys,name_list,color_idx=[0,1,2,3,4],label='cls'):    
    color_shape = (len(gal),len(color_idx))
    pdfc = np.zeros(color_shape)
    pdfz = np.zeros(len(gal),dtype=float)
    pdfm = np.zeros(len(gal),dtype=float)
    pdfr = np.zeros(len(gal),dtype=float)
    
    for idx,name in zip(keys,name_list):
        group = hdf['%s'%(name)]
        ggroup = gal[idx]
        
        pdfr[idx] = interpData(group['pdf_r_%s'%label][:,0],group['pdf_r_%s'%label][:,1],gal['R'][idx])
        pdfz[idx] = interpData(group['pdf_z_%s'%label][:,0],group['pdf_z_%s'%label][:,1],gal['z'][idx])
        pdfc[idx,:] = get_hdf_pdfs(group['pdf_c_%s'%label],ggroup,mode='color',color_idx=color_idx)
        pdfm[idx] = interpData(group['pdf_m_%s'%label][:,0],group['pdf_m_%s'%label][:,1],gal['dmag'][idx])
        
        if label=='field':
            pdfr[idx] = np.median(pdfr[idx])*np.ones_like(pdfr[idx])
    
    pdfc = np.where(pdfc<0.,0.,pdfc)
    
    pdfs = [pdfr,pdfz,pdfc,pdfm]
    return pdfs
```

```python
def plot_histograms(galaxies,prob,axs,xlabel=r'$P_{r}$'):
    mask = galaxies['True'] == True
    nmask = galaxies['True'] == False

    xbins = np.linspace(0.,1.,10)
    
    # axis.hist(prob,bins=xbins,ec='blue', fc='none', lw=1.5, histtype='step',label='All')
    axs.hist(prob[mask],bins=xbins,ec='red', fc='none', lw=1.5, histtype='step',label='True members')
    axs.hist(prob[nmask],bins=xbins,ec='green', fc='none', lw=1.5, histtype='step',label='Non members')
    
    axs.set_xlabel(xlabel,fontsize=16)
    axs.set_ylabel('N')

def plot_single_histogram(galaxies,prob,lcol,title='Buzzard v1.6'):
        
    fig, ax = plt.subplots(figsize=(6,6))
    plot_histograms(galaxies,prob,ax,xlabel=lcol)
    plt.legend(loc='upper left',fontsize=12)
    fig.suptitle(title,fontsize=16)
    
def plot_grid_histograms(galaxies,Pmem_list,title='Buzzard v1.6'):
    fig, axis = plt.subplots(2, 2, figsize=(8,8), sharex='col', sharey='row')

    labels = [['Pr','Pz'],['Pc','Pmem']]
    for i in range(2):
        for j in range(2):
            prob = Pmem_list[i][j]
            plot_histograms(galaxies,prob,axis[i,j],xlabel=labels[i][j])

            if i ==0 and j==0: axis[0,0].legend(loc='upper left',fontsize=12)
    fig.suptitle(title,fontsize=16)
```

```python
def plot_scaling_relation(Pmem,keys,keys2,nbkg,title='Buzzard',original=False):
    ngals = np.array([np.nansum(Pmem[idx]) for idx in keys])
    if original: ngals = np.array([np.nansum(gal['Pmem'][idx]) for idx in keys])
    
    nt = np.array([len(gal2['Pmem'][idx]) for idx in keys2])
    z = cat['redshift']
    linreg=lin_reg(nt,ngals)
        
    idx = np.argsort(nt)
    xt,yh = nt[idx],linreg['Yhat'][idx]

    b0 = round(linreg['b0'],3)
    b1 = round(linreg['b1'],3)
    cb_u, cb_l = linreg['cb_u'], linreg['cb_l']


    fig = plt.figure(figsize=(8,6))
    SNR = nt/np.sqrt(nt+nbkg)
    
    sc = plt.scatter(nt,ngals,c=z,cmap='RdBu',label='$scatter = %.1f$'%(np.std(ngals-nt)))
    plt.plot(xt,yh, color="k",label='y=%.2f+%.2fx'%(b0,b1))
    plt.fill_between(xt, cb_l, cb_u, color="gray", alpha=0.25, label='_nolabel_')
    plt.plot(xt,cb_l, color="gray", label='_nolabel_')
    plt.plot(xt,cb_u, color="gray", label='_nolabel_')

#     plt.scatter(nt,ngals2,alpha=0.2)
    plt.plot(np.linspace(1,800),np.linspace(1,800),linestyle='--',color='k')
    plt.xscale('log')
    plt.yscale('log')
    plt.ylim(2.,800)
    plt.xlim(2.,800)
    plt.xlabel(r'$N_{gals,true}$',fontsize=16)
    plt.ylabel(r'$N_{gals}$',fontsize=16)
    plt.legend(fontsize=12)
    plt.title(title,fontsize=16)
        
    fig.subplots_adjust(right=0.9)
    cbar_ax = fig.add_axes([0.91, 0.15, 0.02, 0.7])
    cb2 = fig.colorbar(sc, cax=cbar_ax, fraction=0.046, pad=0.04)
    cb2.set_label(r'$redshift$',rotation=90)

    
def plot_residuals(Pmem,keys,keys2,nbkg,title='Buzzard'):
    ngals = np.array([np.nansum(Pmem[idx]) for idx in keys])
    ngals2 = np.array([np.nansum(gal['Pmem'][idx]) for idx in keys])
    nt = np.array([len(gal2['Pmem'][idx]) for idx in keys2])
    
    residuals = (ngals-nt)/nt
    
    z = cat['redshift']
    fig = plt.figure(figsize=(8,6))
    SNR = nt/np.sqrt(nt+nbkg)    
    sc = plt.scatter(nt,residuals,c=z,cmap='RdBu',label='$scatter = %.1f$'%(np.std(ngals-nt)))
    plt.axhline(0,color='k',linestyle='--')
    
    plt.ylim(-1.,1.)
    plt.xlim(2.,200)
    plt.xlabel(r'$N_{gals,true}$',fontsize=16)
    plt.ylabel(r'fractional error',fontsize=16)
    plt.legend(fontsize=14)
    plt.title(title,fontsize=16)
    
    fig.subplots_adjust(right=0.9)
    cbar_ax = fig.add_axes([0.91, 0.15, 0.02, 0.7])
    cb2 = fig.colorbar(sc, cax=cbar_ax, fraction=0.046, pad=0.04)
    cb2.set_label(r'$N_{bkg}$',rotation=90)
```
